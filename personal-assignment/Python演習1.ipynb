### 問1

文字列"datascience"の文字を逆に（末尾から先頭に向かって）並べた文字列を出力してください

```python
# 元の文字列
original_string = "datascience"

# スライシング [::-1] を使用して文字列を反転させる。
# [::] は文字列全体を指定することを示す。
# -1 はステップ（文字の取得間隔）を逆方向、すなわち末尾から先頭に向かうよう指定する。
# この操作により、文字列全体が瞬時に反転され、その結果が reversed_string に代入される。
reversed_string = original_string[::-1]

print(reversed_string)
```

### 問2

「ヘエルスシトンニキア」という文字列の1,3,5,7,9文字目を取り出して連結した文字列を出力してください


```python
# 元の文字列を変数 original_string に代入する。
original_string = "ヘエルスシトンニキア"

# スライシング [0::2] を使用して、1文字目から2文字おきに文字を取り出す。
# Pythonのインデックスは 0 から開始されるため、0 が1文字目に相当する。
# スライシングの形式は [開始インデックス : 終了インデックス : ステップ] である。
# 0 は開始位置、2 はステップ（間隔）を指定し、1文字飛ばしで文字を抽出する。
# 終了インデックスを省略することで、文字列の最後まで処理を実行する。
# この操作により、1, 3, 5, 7, 9文字目（ヘ、ル、シ、ン、キ）が取り出され、
# 新しい文字列として連結され、result_string に代入される。
result_string = original_string[0::2]

print(result_string)
```

### 問3

「ヘルシンキ」＋「エストニア」の文字を先頭から交互に連結して文字列「ヘエルスシトンニキア」を出力してください


```python
# 連結する一つ目の文字列
string1 = "ヘルシンキ"

# 連結する二つ目の文字列
string2 = "エストニア"

# zip() 関数と ''.join() メソッドを使用して、二つの文字列を交互に連結する。
# 1. zip(string1, string2) は、各文字列の同じ位置にある文字をペアとして取り出す。
# 2. sum(...) は、リストやタプルなどのイテラブルを展開し、一次元のシーケンスにする。
# 3. ''.join() は、展開された文字のシーケンスを空文字列を区切りとして連結し、最終的な文字列を生成する。
result_string = "".join(sum(zip(string1, string2), ()))

print(result_string)
```

### 問4

"Character education based on the fundamental spirit of Buddhism and the Shiguzeigan"という文を単語に分解し，各単語の（アルファベットの）文字数を先頭から出現順に並べたリストを作成してください


```python
import re

# text に代入
text = "Character education based on the fundamental spirit of Buddhism and the Shiguzeigan"

# 1. 前処理として、re.sub() 関数を用いて文字列から句読点（この例ではコンマ）を取り除く。これにより、単語の文字数を正確に数えることができる。
cleaned_text = re.sub(r'[.,]', '', text)

# 2. cleaned_text.split() メソッドを用いて、文字列を空白で区切り単語のリストに分解する。
words = cleaned_text.split()

# 3. リスト内包表記を使用して、words リストの各単語 (word) に対して len(word) を実行する。len(word) は各単語のアルファベット文字数を返す。
# 結果として得られた文字数のリストが word_lengths に代入される。
word_lengths = [len(word) for word in words]

# 結果の単語リストと文字数リストを出力
print(f"単語リスト{words}")
print(f"各単語の文字数リスト{word_lengths}")
```

### 問5

"It is not the strongest of the species that survives, nor the most intelligent that survives. It is the one that is most adaptable to change."という文を単語に分解し，1, 2, 5, 6, 8, 9, 16番目の単語の先頭の2文字を取り出し，取り出した文字列から単語の位置（先頭から何番目の単語か）を格納した辞書を作成してください

```python
# 英文を変数 text に代入
text = "It is not the strongest of the species that survives, nor the most intelligent that survives. It is the one that is most adaptable to change."

# 単語の位置（1から数える）をリストに代入
target_positions = [1, 2, 5, 6, 8, 9, 16]

# 1. 前処理として、re.sub() 関数を用いて文字列から句読点（コンマ、ピリオド）を取り除く。これにより、単語の文字数を正確に数えることができる。
import re
cleaned_text = re.sub(r'[.,]', '', text)

# 2. cleaned_text.split() メソッドを用いて、文字列を空白で区切り単語のリストに分解する。このリストのインデックスは 0 から始まるため、単語位置との対応に注意が必要である。
words = cleaned_text.split()

# 3. 辞書を格納するための空の辞書を初期化する。
result_dict = {}

# 4. target_positions のリストに含まれる各位置 p についてループ処理を行う。
for p in target_positions:
    # 単語の位置 p は 1 から始まるため、リストのインデックス (i) は p-1 となる。
    word_index = p - 1

    # 該当する位置の単語を取得する。
    # 例: p=1 -> words[0] = "It"
    word = words[word_index]

    # 単語の先頭から2文字を取り出す。
    # スライシング [0:2] は、インデックス 0 から 2 の手前まで（0と1）の文字を取得する。
    first_two_chars = word[0:2]

    # 取得した先頭2文字をキー、元の単語の位置 p を値として辞書に追加する。
    # 例: result_dict["It"] = 1
    result_dict[first_two_chars] = p

# 結果の単語リストと作成した辞書を出力する。
print(f"分解された単語のリスト: {words}")
print(f"作成された辞書: {result_dict}")
```
